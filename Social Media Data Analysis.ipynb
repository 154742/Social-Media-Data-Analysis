```python
import numpy as np 
import pandas as pd 
```

```python
df=pd.read_csv('/kaggle/input/social-media-data-analysis/social_media_data.csv')
```

```python
df
```

```python
df.info()
```

# Replace Data Type Of (date) 

```python
df['date']=pd.to_datetime(df['date'])
```

# Total Number Of Likes, Shares, Comments, And Views 

```python
cat=['likes','shares','comments','views']
df[cat].sum().reset_index()
```

# Highest Engagement

```python
df['engagement'] = df['likes'] + df['shares'] + df['comments']
High_Platform=df.groupby('platform')['engagement'].sum()
High_Platform.idxmax()
```

# Average Number Of Likes, Shares, Comments, And Views

```python
HighPlatform=df.groupby('platform')[cat].mean()#cat=likes,shares,comments,views
HighPlatform
```

# Total Number Of Each Type Of Post

```python
Total_Post=df.groupby('platform')['post_type'].count().reset_index()
Total_Post
```

# Highest Average Engagement

```python
Average_Engagement=df.groupby('post_type')['engagement'].mean()
Average_Engagement.idxmax()
```

#  Total Engagement In Each Quarter Of The Year

```python
df['Quarter']=df['date'].dt.to_period('Q')
df.groupby('Quarter')['engagement'].sum().reset_index()
```
